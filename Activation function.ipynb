## Activation functions.

In Artificial/Deep neural networks, activation functions are crucial components. Due to the need for nonlinearity in neural networks when training models, there has to be functions which transform input to a desired output. This is why Neural networks can learn from complex trends in data problems.
These functions are called activation functions.

Let's explore just 5 of them
1. Rectified linear unit(ReLU)
2. Leaky ReLU
3. Sigmoid
4. Tanh
5. Softmax.

```python
#The 5 activation functions listed above are implemented below using class.

class activations:
   '''
   This is activations class which houses 5 activation functions.
   Input: x. It can be a scaler or an ndarray
   '''
   def __init__(self, x):
      self.x=x

   def relu(self):
       '''
       This is the ReLU activation function.

       Input: There's no argument because the supposed input is the class positional argument

       Output: Returns a scaler or an ndarray
       '''
       return np.maximum(0.0, self.x)

   def leaky_relu(self,alpha=0.01):
      '''
      This is leaky ReLU activation function. It is typically used in the hidden layers.

      Input: alpha. It is a keyword argument.
      Alpha is a hyperparameter what reduces the gradient varnishing problem

      Output: It is either a scaler or an ndarray
      '''
      return np.maximum(alpha*self.x, self.x)

   def sigmoid(self):
      '''
      This is sigmoid function that is used for binary classification. It can also be called logistic function.

      The equation is f(x)=1/(1+exp(-x))

      Input: none. The supposed input is initialized when instantiating the class.
      Output: Either a scaler or an ndarray.
      '''
      return (np.exp(self.x)/(np.exp(self.x)+1))

   def tanh(self):

     '''
     This is hyperbolic tangent function which is typically used in hidden layers of DNN

     The Equation is (e^x-e^-x)/(e^x+e^-x)

     Input: None.
     Output: Either a scaler or an Ndarray.
     '''
     return (np.exp(self.x)-np.exp(-self.x))/(np.exp(self.x)+np.exp(-self.x))

   def softmax(self):
      '''
      This is softmax activation function. It's used for Multi-class classification. This is because of it kind of output. It yields a probabilistic distribution of the input that add up to 1.
      Input: None. It is already initialized in the class instantiation.
      Output: An ndarray
      '''

      U = []
      for i in range(len(self.x)):
         U.append(np.exp(self.x[i])/sum(np.exp(self.x)))
      return np.array(U)
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
F = np.linspace(-10,10,30)
F
```

```python
D =activations(F)
```

```python
D.relu()
```

```python
plt.plot(F, D.relu(), 'r')
plt.title("ReLU function")
plt.xlabel('input value')
plt.ylabel('ReLU')
plt.show()
```

```python
D.leaky_relu(alpha=0.04)
```

```python
plt.plot(F, D.leaky_relu(alpha=0.04), 'g')
plt.title("Leaky_ReLU function")
plt.xlabel('input value')
plt.ylabel('Leaky_ReLU')
plt.show()
```

```python
D.sigmoid()
```

```python
plt.plot(F, D.sigmoid(), 'r')
plt.title("Sigmoid function")
plt.xlabel('input value')
plt.ylabel('sigmoid')
plt.show()
```

```python
D.tanh()
```

```python
plt.plot(F, D.tanh(), 'b')
plt.title("tanh function")
plt.xlabel('input value')
plt.ylabel('tanh')
plt.show()
```

```python
D.softmax()
```

```python
sum(D.softmax())
```
